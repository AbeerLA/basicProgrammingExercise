```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris  # Iris dataset
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Define different values of K
k_values = [4, 5, 6]

# Create subplots for visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, k in enumerate(k_values):
    # Run the K-Means algorithm
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)

    # Plot clustering results
    axes[i].scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

    # Display the cluster centers (centroids)
    centers = kmeans.cluster_centers_
    axes[i].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, label='Centroids')

    axes[i].set_xlabel(iris.feature_names[0])
    axes[i].set_ylabel(iris.feature_names[1])
    axes[i].set_title(f'K-Means Clustering (K={k})')
    axes[i].legend()

# Show all plots
plt.tight_layout()
plt.show()
```
